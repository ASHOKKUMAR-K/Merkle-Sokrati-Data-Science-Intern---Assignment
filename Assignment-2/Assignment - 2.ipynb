```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sb
```

```python
sku = pd.read_excel("data/Data Analyst Assignment.xlsx", sheet_name=3)
sku.head()
```

```python
sku = sku.transpose()
```

```python
sku.columns = sku.iloc[0, :]
```

```python
sku.drop('SKU', inplace=True)
```

```python
sku.head()
```

```python
sku['Month'] = pd.to_datetime(sku.index).month
```

```python
sku['Total Sales'] = sku.iloc[:, :-2].sum(axis = 1)
```

```python
sku.head()
```

# Question 1:

## 1) What is the monthly & quaterly sales data by SKU

```python
month_dict = {8: "August 2019", 9: "September 2019", 10: "October 2019",
             11: "November 2019", 12: "December 2019", 1: "January 2020",
             2 : "February 2020", 3: "March 2020", 4: "April 2020",
             5 : "May 2020", 6: "June 2020", 7: "July 2020"}
months = sku['Month'].unique()
months
```

```python
print("Monthly Sales : ")
monthly_sales = {}
for month in months:
    sales_sum = sku[sku["Month"] == month]['Total Sales'].sum()
    print(month_dict[month], "\t==>\t", sales_sum)
    monthly_sales[month] = sales_sum
```

```python
print("Quartely Sales : \n")
print("Quarter - August 2019 to October 2019 \t:\t ", monthly_sales[8] + monthly_sales[9] + monthly_sales[10])
print("Quarter - November 2019 to January 2020 :\t ", monthly_sales[11] + monthly_sales[12] + monthly_sales[1])
print("Quarter - February 2020 to April 2020 \t:\t ", monthly_sales[2] + monthly_sales[3] + monthly_sales[4])
print("Quarter - May 2020 to July 2020 \t:\t ", monthly_sales[5] + monthly_sales[6] + monthly_sales[7])
```
